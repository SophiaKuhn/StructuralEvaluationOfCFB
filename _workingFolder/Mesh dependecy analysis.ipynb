```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## 1.  Data Import 

```python
#Mesh size 1
df_1 = pd.read_csv("trainingData/1001_CFB_xy.csv")
#Mesh size 3
df_3 = pd.read_csv("trainingData/1000_CFB_xy.csv")
```

## 2. Compare the 2 datasets

### 2.1 Diagonal plot

```python
cm = 1/2.54

fig, axes = plt.subplots(1,3,figsize=(75*cm, 15*cm))

axes[0].scatter(df_1["eta_min_s"], df_3["eta_min_s"], label=r"$\eta_{min,s}$", marker="o", s=10, c="gray")
axes[0].plot(df_1["eta_min_s"], df_1["eta_min_s"], label="Perfect correspondance", color = "black", linestyle = "--")
axes[0].set_xlabel("Mesh size 1")
axes[0].set_ylabel("Mesh size 3")
axes[0].grid()
axes[0].legend(loc='upper left', bbox_to_anchor=(1, 1))
axes[0].axis("square")

axes[1].scatter(df_1["eta_min_c"], df_3["eta_min_c"], label=r"$\eta_{min,c}$", marker="o", s=10, c="gray")
axes[1].plot(df_1["eta_min_c"], df_1["eta_min_c"], label="Perfect correspondance", color = "black", linestyle = "--")
axes[1].set_xlabel("Mesh size 1")
axes[1].set_ylabel("Mesh size 3")
axes[1].grid()
axes[1].legend(loc='upper left', bbox_to_anchor=(1, 1))
axes[1].axis("square")

axes[2].scatter(df_1["eta_min_shear"], df_3["eta_min_shear"], label=r"$\eta_{min,shear}$", marker="o", s=10, c="gray")
axes[2].plot(df_1["eta_min_shear"], df_1["eta_min_shear"], label="Perfect correspondance", color = "black", linestyle = "--")
axes[2].set_xlabel("Mesh size 1")
axes[2].set_ylabel("Mesh size 3")
axes[2].grid()
axes[2].legend(loc='upper left', bbox_to_anchor=(1, 1))
axes[2].axis("square")

fig.tight_layout()
```

### 2.5 Relative difference

#### 2.5.1 Remove lines with zeros as they don't make much sense

```python
df_analysis = pd.DataFrame()

df_analysis["mesh_size_1_eta_min_s"] = df_1["eta_min_s"]
df_analysis["mesh_size_1_eta_min_c"] = df_1["eta_min_c"]
df_analysis["mesh_size_1_eta_min_shear"] = df_1["eta_min_shear"]

df_analysis["mesh_size_3_eta_min_s"] = df_3["eta_min_s"]
df_analysis["mesh_size_3_eta_min_c"] = df_3["eta_min_c"]
df_analysis["mesh_size_3_eta_min_shear"] = df_3["eta_min_shear"]
```

```python
df_analysis
```

```python
df_analysis.drop(df_analysis[df_analysis.isin([0.0]).any(axis=1)].index, inplace = True)
```

```python
df_analysis
```

#### 2.5.2 Remove lines with eta smaller than 0.5 (in mesh size 1)

```python
df_analysis.drop(df_analysis[df_analysis["mesh_size_1_eta_min_s"] < 0.5].index, inplace = True)
df_analysis.drop(df_analysis[df_analysis["mesh_size_1_eta_min_c"] < 0.5].index, inplace = True)
df_analysis.drop(df_analysis[df_analysis["mesh_size_1_eta_min_shear"] < 0.5].index, inplace = True)
```

```python
df_analysis
```

#### 2.5.3 Plot the now shrunken data

```python
cm = 1/2.54

fig, axes = plt.subplots(1,3,figsize=(75*cm, 15*cm))

axes[0].scatter(df_analysis["mesh_size_1_eta_min_s"], 100*(df_analysis["mesh_size_3_eta_min_s"] - df_analysis["mesh_size_1_eta_min_s"])/df_analysis["mesh_size_1_eta_min_s"], marker="o", s=10, c="gray")
axes[0].set_xlabel(r"$\eta_{min,s}$")
axes[0].set_ylabel("Relative Difference [%]")
axes[0].grid()

axes[1].scatter(df_analysis["mesh_size_1_eta_min_c"], 100*(df_analysis["mesh_size_3_eta_min_c"] - df_analysis["mesh_size_1_eta_min_c"])/df_analysis["mesh_size_1_eta_min_c"], marker="o", s=10, c="gray")
axes[1].set_xlabel(r"$\eta_{min,c}$")
axes[1].set_ylabel("Relative Difference [%]")
axes[1].grid()

axes[2].scatter(df_analysis["mesh_size_1_eta_min_shear"], 100*(df_analysis["mesh_size_3_eta_min_shear"] - df_analysis["mesh_size_1_eta_min_shear"])/df_analysis["mesh_size_1_eta_min_shear"], marker="o", s=10, c="gray")
axes[2].set_xlabel(r"$\eta_{min,shear}$")
axes[2].set_ylabel("Relative Difference [%]")
axes[2].grid()

fig.tight_layout()
```

### 2.6 Delve deeper in the region of interest

```python
df_analysis
```

#### 2.6.1 Get only data in region of interest

```python
df_eta_s = df_analysis.drop(df_analysis[df_analysis["mesh_size_1_eta_min_s"] > 10.0].index)[["mesh_size_1_eta_min_s", "mesh_size_3_eta_min_s"]]
df_eta_c = df_analysis.drop(df_analysis[df_analysis["mesh_size_1_eta_min_c"] > 10.0].index)[["mesh_size_1_eta_min_c", "mesh_size_3_eta_min_c"]]
df_eta_shear = df_analysis.drop(df_analysis[df_analysis["mesh_size_1_eta_min_shear"] > 10.0].index)[["mesh_size_1_eta_min_shear", "mesh_size_3_eta_min_shear"]]
```

```python
df_eta_s["diff_eta_min_s"] = df_eta_s["mesh_size_3_eta_min_s"] - df_eta_s["mesh_size_1_eta_min_s"]
df_eta_s["rel_diff_eta_min_s"] = 100*(df_eta_s["mesh_size_3_eta_min_s"] - df_eta_s["mesh_size_1_eta_min_s"])/df_eta_s["mesh_size_1_eta_min_s"]

df_eta_c["diff_eta_min_c"] = df_eta_c["mesh_size_3_eta_min_c"] - df_eta_c["mesh_size_1_eta_min_c"]
df_eta_c["rel_diff_eta_min_c"] = 100*(df_eta_c["mesh_size_3_eta_min_c"] - df_eta_c["mesh_size_1_eta_min_c"])/df_eta_c["mesh_size_1_eta_min_c"]

df_eta_shear["diff_eta_min_shear"] = df_eta_shear["mesh_size_3_eta_min_shear"] - df_eta_shear["mesh_size_1_eta_min_shear"]
df_eta_shear["rel_diff_eta_min_shear"] = 100*(df_eta_shear["mesh_size_3_eta_min_shear"] - df_eta_shear["mesh_size_1_eta_min_shear"])/df_eta_shear["mesh_size_1_eta_min_shear"]
```

#### 2.6.2 Mean difference

```python
cm = 1/2.54

fig, axes = plt.subplots(1,3,figsize=(75*cm, 15*cm))

bins_s = np.linspace(0.5,10.0,20)
group_s = df_eta_s.groupby(pd.cut(df_eta_s.mesh_size_1_eta_min_s, bins_s))
plot_centers_s = (bins_s[:-1] + bins_s[1:])/2
plot_values_s = group_s.diff_eta_min_s.mean()
axes[0].plot(plot_centers_s, plot_values_s)
axes[0].set_xlabel(r"$\eta_{min,s}$")
axes[0].set_ylabel("Difference [-]")
axes[0].grid()
axes[0].set_xlim(left=0.5,right=10.0)
axes[0].set_ylim(bottom=0.0,top=10.0)
axes[0].locator_params(axis='x', nbins=20)
axes[0].locator_params(axis='y', nbins=20)

bins_c = np.linspace(0.5,10.0,20)
group_c = df_eta_c.groupby(pd.cut(df_eta_c.mesh_size_1_eta_min_c, bins_c))
plot_centers_c = (bins_c[:-1] + bins_c[1:])/2
plot_values_c = group_c.diff_eta_min_c.mean()
axes[1].plot(plot_centers_c, plot_values_c)
axes[1].set_xlabel(r"$\eta_{min,c}$")
axes[1].set_ylabel("Difference [-]")
axes[1].grid()
axes[1].set_xlim(left=0.5,right=10.0)
axes[1].set_ylim(bottom=0.0,top=5.0)
axes[1].locator_params(axis='x', nbins=20)
axes[1].locator_params(axis='y', nbins=10)

bins_shear = np.linspace(0.5,10.0,20)
group_shear = df_eta_shear.groupby(pd.cut(df_eta_shear.mesh_size_1_eta_min_shear, bins_shear))
plot_centers_shear = (bins_shear[:-1] + bins_shear[1:])/2
plot_values_shear = group_shear.diff_eta_min_shear.mean()
axes[2].plot(plot_centers_shear, plot_values_shear)
axes[2].set_xlabel(r"$\eta_{min,shear}$")
axes[2].set_ylabel("Difference [-]")
axes[2].grid()
axes[2].locator_params(axis='x', nbins=20)
axes[2].set_xlim(left=0.5,right=10.0)
axes[2].set_ylim(bottom=0.0,top=5.0)
axes[2].locator_params(axis='x', nbins=20)
axes[2].locator_params(axis='y', nbins=10)

fig.tight_layout()
```
