```python
import pickle
import pandas as pd
import os
import matplotlib
import math as m
```

```python
#Util functions
def create_folder_if_not_exists(folder_path):
    if not os.path.exists(folder_path):
        try:
            os.makedirs(folder_path)
            print(f"Folder '{folder_path}' created successfully.")
        except OSError as e:
            print(f"Error creating folder '{folder_path}': {e}")
```

## Retrieve results

### mesh size 1

```python
# NEW: Eingabe
idx_s1=77
start_id = 0
end_id = 14
step='step_4'
var_par='s_walls'
unit=' [mm]'

df_xy_1=None

#current_directory = os.getcwd()

xy_file_name='CFBData\\{}_Batch\\{}_CFB_xy.csv'.format(idx_s1,idx_s1)

df_xy_1=pd.read_csv(xy_file_name)
df_xy_1[[var_par,'eta_min_s','x_s','y_s','z_s','Location_s','GP_count_s','eta_min_c','x_c','y_c','z_c','Location_c','GP_count_c']]
```

### mesh size 3

```python
# NEW: Eingabe
idx_s3=75
start_id = 0
end_id = 14
step='step_4'
var_par='s_walls'
unit=' [mm]'

df_xy_3=None

#current_directory = os.getcwd()

xy_file_name='CFBData\\{}_Batch\\{}_CFB_xy.csv'.format(idx_s3,idx_s3)

df_xy_3=pd.read_csv(xy_file_name)
df_xy_3[[var_par,'eta_min_s','x_s','y_s','z_s','Location_s','GP_count_s','eta_min_c','x_c','y_c','z_c','Location_c','GP_count_c']]
```

## Only mesh size 1

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.plot(df_xy_1[var_par], df_xy_1['eta_min_s'],label='Steel Reinforcement (ms=1t)', marker='o', color='darkblue')
ax.plot(df_xy_1[var_par], df_xy_1['eta_min_c'], label='Concrete (ms=1t)', marker='x', color='green')

ax.set(xlabel=var_par+unit, ylabel='$\eta \ min$ [-]',
       title='Concrete Frame Bridge, Studie {}'.format(idx_s2))

ax.set_ylim(bottom=0,top=20)
ax.axhline(y=1, color='red', linestyle='--', linewidth=1, label='$\mu=1$')

ax.grid()
ax.legend()

#fig.savefig("test.png")
# plt.show()
```

```python
# save figure
figname='eta_plot_var_{}'.format(var_par)
fig_folder_path='CFBData\\{}_Batch\\_figures'.format(idx_s2)
create_folder_if_not_exists(fig_folder_path)
fig.savefig(fig_folder_path+"\\"+figname+".png")
```

```python
import matplotlib.pyplot as plt

cm = 1/2.54

fig, ax = plt.subplots(figsize=(25*cm,15*cm))
ax.plot(df_xy_3[var_par], df_xy_3['eta_min_s'],label='Steel Reinforcement (ms=3t)', marker='o', color='lightblue')
ax.plot(df_xy_3[var_par], df_xy_3['eta_min_c'], label='Concrete (ms=3t)', marker='x', color='lightgreen')

ax.plot(df_xy_1[var_par], df_xy_1['eta_min_s'],label='Steel Reinforcement (ms=1t)', marker='o', color='darkblue')
ax.plot(df_xy_1[var_par], df_xy_1['eta_min_c'], label='Concrete (ms=1t)', marker='x', color='green')

ax.set(xlabel=var_par+unit, ylabel='$\eta \ min$ [-]',
       title='Concrete Frame Bridge, Studie {} and {}'.format(idx_s1,idx_s3))

ax.set_ylim(bottom=0,top=20)
ax.axhline(y=1, color='red', linestyle='--', linewidth=1, label='$\mu=1$')

ax.grid()
ax.legend(loc='upper left', bbox_to_anchor=(1, 1))
fig.tight_layout()
 
#fig.savefig("test.png")
# plt.show()
```
